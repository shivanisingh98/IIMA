```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import requests
import StringIO
import zipfile
import scipy.stats
```

```python
rf=pd.read_csv("/home/shivani/data_csv/peak_red_fridge.csv")
ac=pd.read_csv("/home/shivani/data_csv/peak_red_ac.csv")
col=pd.read_csv("/home/shivani/data_csv/peak_red_cooler.csv")
wm=pd.read_csv("/home/shivani/data_csv/peak_red_wm.csv")
```

```python
fig, ax = plt.subplots()
ax.scatter(rf.ratio,rf.red)

fit = np.polyfit(rf.ratio,rf.red, deg=1)
print fit
ax.scatter(rf.ratio,rf.red,s=10,c='blue')
ax.plot(rf.ratio, fit[0] * rf.ratio+ fit[1], color='orange')
plt.title("Percentage Peak reduction for Refrigerator",y=1.04)
plt.ylabel("Percentage Peak reduction")
plt.xlabel("Peak price/Off-Peak Price")
fig.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(ac.ratio,ac.red)

fit = np.polyfit(ac.ratio,ac.red, deg=1)
print fit
ax.scatter(ac.ratio,ac.red,s=10,edgecolors='black',c='black')

ax.plot(ac.ratio, fit[0] * ac.ratio+ fit[1], color='orange')
plt.title("Percentage Peak reduction for Air Conditioner",y=1.04)
plt.ylabel("Percentage Peak reduction")
plt.xlabel("Peak price/Off-Peak Price")
fig.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(col.ratio,col.red)

fit = np.polyfit(col.ratio,col.red, deg=1)
print fit
ax.scatter(col.ratio,col.red,s=10,c='green')
plt.title("Percentage Peak reduction for Cooler",y=1.04)
plt.ylabel("Percentage Peak reduction")
plt.xlabel("Peak price/Off-Peak Price")
ax.plot(col.ratio, fit[0] * col.ratio+ fit[1], color='orange')

fig.show()
```
