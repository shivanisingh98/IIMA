```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import StringIO
import zipfile
import scipy.stats
```

```python
ac=pd.read_csv("/home/shivani/data_csv/middle_ac.csv")
print ac.shape
ac.head()

y1=list(ac['base-2'])+list(ac['base-1'])+list(ac['base'])+list(ac['base+2'])+list(ac['base+5'])+list(ac['base+10'])
l=len(ac)
x1=[1]*l+[2]*l+[3]*l+[4]*l+[5]*l+[6]*l
y1=pd.Series(y1)
x1=pd.Series(x1)

dbac=pd.DataFrame([x1,y1])
dbac=dbac.transpose()
dbac.columns=['index','use']


```

```python

fig, ax = plt.subplots()
ax.scatter(dbac['index'],dbac.use)

fit = np.polyfit(dbac['index'],dbac.use, deg=1)
print fit
ax.scatter(dbac['index'],dbac.use,s=10,edgecolors='black',c='black')

ax.plot(dbac['index'], fit[0] * dbac['index']+ fit[1], color='orange')
plt.title("Air Conditioner Usage variation with price of electricity",y=1.04)
plt.ylabel("Number of units consumed daily")
plt.xticks(np.arange(1,7,1),['Base Price-Rs.4','Base Price-Rs.2','Base Price','Base Price+Rs.2','Base Price+Rs.5',
                          'Base Price+Rs.10'],rotation=90)
fig.show()
```
